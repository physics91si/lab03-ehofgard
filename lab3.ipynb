```python
# Don't edit this function
def load_sample():
    """Return the entire text of Hamlet in a string."""
    with open('hamlet.txt') as f:
        sample = f.read()
    return sample
```

```python
# Edit this function. "pass" tells Python to do nothing.

from string import punctuation

def strip_punctuation(s):
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    newstring = ''
    for char in s:
        if char not in punctuations:
            newstring = newstring + char
    return newstring

def parse_sample(text, words_only, sort_list):
    wordlist = text.split()
    
    newstring = ""
    if words_only == True:
        for i in range(0, len(wordlist)):
            word = wordlist[i].lower()
            word = strip_punctuation(word)
            wordlist[i] = word 
    if sort_list == True:
        wordlist.sort()
    return wordlist            
            
    
```

```python
text = load_sample()
print (parse_sample(text, True, True))
```

```python
import math
from __future__ import division
## The numpy library could also be used to write many of these functions
import numpy as np

def count_freq(lst):
    
    freq_dict = dict()
    for word in lst:
        count = 1
        if word not in freq_dict:
            freq_dict[word] = count
        else:
            count = count + 1
            freq_dict[word] = count
    return freq_dict

def mean(lst):
    count = len(lst)
    total = 0
    for number in lst:
        total = total + number
    mean = total / count
    return mean

def std_dev(lst):
    data_mean = mean(lst)
    total = len(lst)
    std_dev_sum = 0
    for number in lst:
        std_dev_sum = std_dev_sum + (number - data_mean)**2
    variance = (1/total)*std_dev_sum
    std_dev = math.sqrt(variance)
    return std_dev

## Used prior knowledge of python libraries to write this in a one line function
def median(lst):
   return np.median(lst)
    
```

```python
text = load_sample()
lst = (parse_sample(text, True, True))

print count_freq(lst)
print mean(lst)
```

```python
def print_stats(lst):
    print('This will print the mean, standard deviation, and median of a given dataset.')
    print('Mean: ' + str(mean(lst)))
    print('Standard Deviation: ' + str(std_dev(lst)))
    print('Median: ' + str(median(lst)))
```
